# K Mean Clustering
* K mean clustering is unsupervised learning algorithm.K represents no of clusters
* It always try to keep points in a group as similar as possible and points from different group as dissimilar as possible
* k-Means clustering is a centriod based clustering algorithm 

```python
#importing libraries 
#pandas for loading data and manipulation
#numpy for working with multidimensional arrays and matrices
#searborn and matplotlib is used for data visualization
#sklearn is machine learning library where all the methods related to ML algorithms can be imported


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline

#X contains random values with shape(100,2)
X= -2 * np.random.rand(100,2)
#X1 contains positive random values with shape(50,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1 #A total of 100 data points has been generated and divided into two groups, of 50 points each.
plt.scatter(X[ : , 0], X[ :, 1], s = 50, c = 'b') #scattering the points on positive values and negative values
plt.show()
```

```python
kmeans=KMeans(n_clusters=2) #calling kmeans constructor with n_clusters=2 means with two groupings or clusters
kmeans.fit(X) 
```

```python
kmeans.cluster_centers_ #Finding the centriods of each cluster
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')#Plotting the points on the  x and y axis
plt.scatter(-0.94665068, -0.97138368, s=200, c='g', marker='s') #plotting centroids 
plt.scatter(2.01559419, 2.02597093, s=200, c='r', marker='s')
plt.show()
```

```python
# Testing the algorithm to predict the values (-1,-3) to know which clusters it belongs to
sample_test=np.array([-1.0,-3.0])
second_test=sample_test.reshape(1, -1)
kmeans.predict(second_test)
```

* we can see that points -1,-3 belongs to class 0 which means negative class

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(-0.94665068, -0.97138368, s=200, c='g', marker='s')
plt.scatter(2.01559419, 2.02597093, s=200, c='r', marker='s')
plt.plot(-1.0,-3.0,'r+')
plt.show()
```
